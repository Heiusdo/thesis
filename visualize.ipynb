```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Training history
epochs = [1, 2, 3]
train_loss = [0.709991, 0.626618, 0.573117]
train_acc = [0.6932, 0.7329, 0.7618]
```

```python
metrics = {
    "Accuracy": 0.721716,
    "Precision (macro)": 0.721137,
    "Recall (macro)": 0.723189,
    "F1-score (macro)": 0.721588,
    "ROC-AUC": 0.879123
}
```

```python
plt.figure()
plt.plot(epochs, train_loss, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("RoBERTa Training Loss")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(epochs, train_acc, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Training Accuracy")
plt.title("RoBERTa Training Accuracy")
plt.grid(True)
plt.show()
```

```python
names = list(metrics.keys())
values = list(metrics.values())

plt.figure()
plt.bar(names, values)
plt.ylim(0, 1)
plt.ylabel("Score")
plt.title("RoBERTa Evaluation Metrics")
plt.xticks(rotation=30)
plt.tight_layout()
plt.grid(axis="y")
plt.savefig("pic/roberta_finetuned_metrics.png")
plt.show()
```
